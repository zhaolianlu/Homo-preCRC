```r
#set path
setwd('D:/2.crc_trio/scCrypts/result/')
suppdir="D:/2.crc_trio/scCrypts/result/plot/0.supp/"
maindir="D:/2.crc_trio/scCrypts/result/plot/"

#load library
library(ggtree)
library(treeio)
library(ggplot2)
library(dplyr)
library(stringr)
```

```r
#读入newick格式的进化树文件；
tree = read.newick("D:/2.crc_trio/scCrypts/result/runtime/VariantFilter/B139_SNP_vd7_vr4.nj_rescale.nwk")
tree2=drop.tip(tree, 'ref')
df_info  = data.frame(sn=tree2$tip.label) %>% mutate(region=ifelse(grepl('N_',sn),substr(sn,5,5),substr(sn,5,6)))

g1=tree2 %>% ggtree(.,layout = "rectangular",branch.length='branch.length')+
geom_tiplab(size=2,align=TRUE,linesize=.5,color="grey10") 
g1<-g1 %<+% df_info  + geom_tippoint(aes(color=region))+ 
  theme(legend.position="right")+theme_tree2()+ xlab('Branch length in mutation number')

print(g1)
ggsave(paste0(suppdir,"fig3.sgWGS_B139_ggtree5.pdf"),width=5, height=6)
dev.off()
```
